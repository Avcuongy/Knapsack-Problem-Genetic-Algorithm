# Tình huống tổng quan: 
Các bài toán Knapsack trong 6 test case có số lượng vật phẩm và công suất tối đa khác nhau. Các thuật toán di truyền sử dụng quần thể có tỷ lệ lai ghép từ 0.7 đến 0.9, và các yếu tố đo lường gồm:

- Fitness cuối cùng: Giá trị tối ưu của quần thể sau khi chạy thuật toán.
- Tốc độ hội tụ: Số thế hệ cần thiết để đạt được kết quả tối ưu.
- Độ ổn định (std): Sự thay đổi trong Fitness giữa các cá thể trong quần thể.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from Genetic_Algorithm import * 
```

```python
items, max_capacity = get_user_input()
get_info()
```

### Ghi lại thông số thay đổi với từng tỷ lệ lai ghép

```python
# Thực hiện thử nghiệm với các tỷ lệ lai ghép khác nhau
crossover_rates = [0.7, 0.8, 0.9]
experiment_results = {}

def run_experiment_with_crossover(cros_rate):
    global CROSSOVER_RATE, fitness_history
    
    # Lưu giá trị CROSSOVER_RATE ban đầu
    original_cros_rate = CROSSOVER_RATE
    
    # Thay đổi CROSSOVER_RATE cho thử nghiệm
    CROSSOVER_RATE = cros_rate
    
    # Thực hiện thuật toán di truyền
    best_solution, best_fitness = genetic_algorithm()
    
    # Lưu kết quả
    result = {
        'best_solution': best_solution,
        'best_fitness': best_fitness,
        'fitness_history': fitness_history.copy()  # Lưu lại lịch sử fitness
    }
    
    # Reset fitness_history sau mỗi lần thử nghiệm
    fitness_history.clear()
    
    # Khôi phục giá trị CROSSOVER_RATE ban đầu
    CROSSOVER_RATE = original_cros_rate
    
    return result

# Thực hiện các thử nghiệm cho các tỷ lệ lai ghép khác nhau
for cros_rate in crossover_rates:
    print(f"\nThử nghiệm với tỷ lệ lai ghép: {cros_rate}")
    
    # Chạy thử nghiệm
    result = run_experiment_with_crossover(cros_rate)
    
    # Lưu kết quả vào experiment_results
    experiment_results[cros_rate] = result
    
    print(f"Best Solution: {result['best_solution']}")
    print(f"Best Fitness: {result['best_fitness']}")
    print(f"Fitness History: {result['fitness_history']}")
```

### Fitness qua các thế hệ với thông số tỷ lệ lai ghép khác nhau

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Lặp qua các tỷ lệ lai ghép và vẽ biểu đồ
for crossover_rate in crossover_rates:
    plt.plot(experiment_results[crossover_rate]['fitness_history'], 
             label=f'Crossover Rate = {crossover_rate:.2f}')

# Thiết lập nhãn, tiêu đề và hiển thị biểu đồ
plt.xlabel('Generation')
plt.ylabel('Best Fitness Value')
plt.title('So sánh fitness qua các thế hệ với các tỷ lệ lai ghép khác nhau')
plt.legend()
plt.grid(True)
plt.show()
```

### Bảng kết quả

| Crossover Rate                | Best Fitness                                              | Convergence Speed                                            | Stability (Std)                  |
|----------------------------------|-----------------------------------------------------------|-------------------------------------------------------------|-----------------------------------|
| Tỷ lệ lai ghép (Phần trăm cá thể trong quần thể sẽ được lai ghép)    | Fitness tốt nhất trong quần thể tại một thế hệ nhất định | Tốc độ hội tụ (số thế hệ hoặc thời gian để đạt mức tối ưu)  | Độ ổn định trung bình qua các thế hệ |


```python
# Hiển thị bảng kết quả chi tiết
print("\nKết quả chi tiết:")
print("-" * 80)
print(f"| {'Crossover Rate':^15} | {'Best Fitness':^15} | {'Convergence Speed':^20} | {'Stability (Std)':^17} |")
print("-" * 80)

for crossover_rate in crossover_rates:
    history = experiment_results[crossover_rate]["fitness_history"]
    final_fitness = history[-1]
    
    # Tính tốc độ hội tụ (số thế hệ để đạt 95% fitness cuối cùng)
    convergence_gen = next(
        (i for i, x in enumerate(history) if x >= 0.95 * final_fitness), len(history)
    )
    
    # Tính độ ổn định (độ lệch chuẩn qua các thế hệ)
    stability_std = np.std(history)

    print(f"| {crossover_rate:^15.2f} | {final_fitness:^15.2f} | {convergence_gen:^20} | {round(stability_std, 2):^17.2f} |")

print("-" * 80)
```

### Phân tích độ ổn định của fitness qua các thế hệ

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(12, 6))

# Lặp qua các tỷ lệ lai ghép và tính độ ổn định
for crossover_rate in crossover_rates:
    history = experiment_results[crossover_rate]['fitness_history']
    
    # Tính độ dao động (rolling standard deviation)
    rolling_std = pd.Series(history).rolling(window=10).std()
    
    # Vẽ biểu đồ
    plt.plot(rolling_std, label=f'Crossover Rate = {crossover_rate:.2f}')

# Thiết lập nhãn và tiêu đề
plt.xlabel('Generation')
plt.ylabel('Standard Deviation (Rolling Window = 10)')
plt.title('Độ ổn định của fitness qua các thế hệ với các tỷ lệ lai ghép khác nhau')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Tổng kết kết quả cho từng tỷ lệ lai ghép
for crossover_rate in crossover_rates:
    history = experiment_results[crossover_rate]['fitness_history']
    final_fitness = history[-1]
    
    # In ra kết quả tổng kết
    print(f"\nTỷ lệ lai ghép {crossover_rate:.2f}:")
    print(f"- Fitness cuối cùng: {final_fitness}")
    
    # Tính tốc độ hội tụ: số thế hệ để đạt ít nhất 95% fitness cuối cùng
    convergence_gen = next((i for i, x in enumerate(history) if x >= 0.95 * final_fitness), len(history))
    print(f"- Tốc độ hội tụ: {convergence_gen} thế hệ")
    
    # Tính độ ổn định (độ lệch chuẩn qua các thế hệ)
    stability_std = np.std(history)
    print(f"- Độ ổn định (std): {stability_std:.2f}")
```
